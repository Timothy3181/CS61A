表达式：<br>
使用函数调用符号来写下所有的东西<br>
正常的运算：

```python
# add
2000 + 15
```

多个数组连续计算：

```python
1 * 2 * ((3 * 4 * 5 // 6) ** 3) + 7 + 8
```

函数表达式的调用有返回值：

```python
max(2, 4)
```

```python
min(-2, 500)
```

尽管可以用+, *来进行加法和乘法，仍然有函数可以进行这些操作

```python
from operator import add
add(2, 4)
```

```python
from operator import mul
mul(2, 4)
```

表达式是可以嵌套的，有些函数表达式甚至可以有多参数

```python
max(1, 2, 3, 4, 5)
```

```python
from operator import add, mul
add(max(1, 2, 3, 4, 5), add(3, mul(add(2, 4), 2)))
```

表达式的组成：<br>
`add(2, 3)`<br>
从结构上说<br>
1.表达式总有括号存在<br>
2.整个东西都是一个调用表达式，其中包括其他表达式<br>
3.开括号前的所有内容（开括号就是'(', '[', '{'）都被称为运算符子表达式<br>
··这里的运算符子表达式其实可能就是把这个表达式看成一个更大的表达式，比如`add(add(2, 3), 4)`这里，后一个add就很明显是一个子表达式，因为它被嵌套在了外面的add函数调用表达式中，所以它是一个子表达式，而外面的add函数调用表达式实际上也是一个子表达式，因为它能返回一个值，当这个函数在调用的时候被写成`result = add(add(2, 3), 4)`时，就更加明显了，总的来说，这里只是在强调一种嵌套的形式<br>
4.操作数(operands)在括号内由逗号分隔，运算符和操作数也是表达式
5.最终表达式会被计算为一个值<br>
从调用表达式上说<br>
1.调用表达式的评估方式相同，这是一个评估过程（现在，编程语言的工作方式是通过应用评估过程来解释表达式，它们只是按照程序一遍又一遍地应用相同的过程）<br>
2.评估过程顺序如下<br>
（1）按顺序评估运算符，然后评估操作数子表达式<br>
（2）然后从运算符中获取一个函数，从操作数中获取参数，然后将该运算符的值作为函数应用于操作数的值。<br>
··主要要注意评估的顺序
